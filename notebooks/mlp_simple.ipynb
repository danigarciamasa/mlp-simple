```python
import tensorflow as tf
import numpy as np
import pandas as pd
import time
print("TensorFlow:", tf.__version__)
```

```python
from tensorflow import keras

(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
X_train.shape, X_test.shape
```

```python
X_train, y_train = X_train[:10000], y_train[:10000]
X_test, y_test = X_test[:2000], y_test[:2000]
```

```python
X_train = X_train.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0
X_train = X_train.reshape((-1, 28*28))
X_test = X_test.reshape((-1, 28*28))
```

```python
from tensorflow.keras import layers
model = keras.Sequential([
layers.Input(shape=(784,)),
layers.Dense(64, activation="relu"),
layers.Dense(10, activation="softmax")
])
model.summary()
```

```python
model.compile(
optimizer="adam",
loss="sparse_categorical_crossentropy",
metrics=["accuracy"]
)
```

```python
history = model.fit(
X_train, y_train,
validation_split=0.1,
epochs=5,
batch_size=128
)
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(test_loss, test_acc)
```

```python
model.save("/content/mlp_model.h5")
```

```python
summary = []
model.summary(print_fn=lambda x: summary.append(x))
summary = "\n".join(summary)
report = f"""
## Informe del experimento MLP
### Arquitectura
{summary}

### MÃ©tricas finales
- Test loss: {test_loss}
- Test accuracy: {test_acc}

### Fecha
{time.ctime()}
"""

# Crear archivo report.md con el informe
with open("/content/report.md", "w") as f:
    f.write(report)
```

```python
# ============================
# ğŸ”’ VerificaciÃ³n de archivos MLP
# ============================

import hashlib

# FunciÃ³n para calcular hash SHA-256
def sha256(path):
    h = hashlib.sha256()
    with open(path, "rb") as f:
        h.update(f.read())
    return h.hexdigest()

# Archivos generados por el experimento
hash_model = sha256("/content/mlp_model.h5")     # modelo entrenado
hash_report = sha256("/content/report.md")       # informe del experimento

# Mostrar resultados
print("âœ… VerificaciÃ³n completada:\n")
print("ğŸ“¦ Hash del modelo (.h5):", hash_model)
print("ğŸ“ Hash del informe (.md):", hash_report)

# Guardar los hashes en un archivo resumen
hash_summary = f"""
# VerificaciÃ³n de archivos MLP Simple
**Modelo (.h5)**
`{hash_model}`

**Informe (.md)**
`{hash_report}`

Generado automÃ¡ticamente en Colab.
"""

with open("/content/hash_summary.md", "w") as f:
    f.write(hash_summary)

print("\nğŸ“ Archivo 'hash_summary.md' creado correctamente.")
```

**ğŸ”¹ QuÃ© he entendido del MLP:**  
El MLP (PerceptrÃ³n Multicapa) es una red neuronal que aprende a reconocer patrones ajustando los pesos de sus conexiones a travÃ©s del entrenamiento.  
Funciona mediante capas de neuronas conectadas entre sÃ­: una capa de entrada que recibe los datos, una o varias capas ocultas que aprenden las relaciones no lineales y una capa de salida que produce la predicciÃ³n.  
He comprendido que cuanto mÃ¡s capas y neuronas tiene, mayor es su capacidad de aprendizaje, pero tambiÃ©n aumenta el riesgo de **overfitting** si no se regula bien.

---

**ğŸ”¹ QuÃ© significa reproducibilidad:**  
La reproducibilidad consiste en poder repetir un experimento y obtener los mismos resultados utilizando el mismo cÃ³digo, los mismos datos y la misma configuraciÃ³n.  
En el contexto de redes neuronales, esto permite verificar que los resultados no son producto del azar y que el proceso de entrenamiento es confiable y transparente.

---

**ğŸ”¹ QuÃ© aporta registrar hashes:**  
Registrar **hashes** garantiza la **integridad** y **autenticidad** de los archivos generados (modelo, informe, notebook).  
El hash actÃºa como una huella digital Ãºnica: si un archivo cambia aunque sea un solo byte, su hash tambiÃ©n cambia.  
De este modo, se puede comprobar que los resultados del experimento no han sido modificados y que el trabajo es totalmente **reproducible y verificable**.
